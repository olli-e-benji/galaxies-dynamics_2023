## Initial Data
To better understand the choice of our initial datas, we first have to show the conversion we used to go from Internal Units to the cgs system.<br>
From the definition of *centripetal force*<br><br>
$$ f_\text{cp} = \dfrac{v^2}{r} = \dfrac{GM}{r^2}$$
<br>
we define the following *adimensional quantity* for our conversion
<br><br>
$$ \bigg(\dfrac{GM}{rv^2}\bigg)_\text{iu} = \bigg(\dfrac{GM}{rv^2}\bigg)_\text{cgs} $$
<br>
Since the quantity is derived from the centripetal force we decided to use the mass ratios and distance from the Earth-Sun system, interpreting $m_\text{iu} = 1 = 1\ M_\odot$. From which

$$
\dfrac{m_\text{cgs}}{m_\text{iu}} = 2 \times 10^{33} \qquad \dfrac{r_\text{cgs}}{r_\text{iu}} = 1.49 \times 10^{13}
$$

From which we find
$$
\dfrac{v_\text{cgs}}{v_\text{iu}} = 3 \times 10^6\
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm

#config plot
if "setup_text_plots" not in globals():
    from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=12, usetex=True)
%config InlineBackend.figure_format='retina'
```

We generated 300 point-like particles of equal mass $m=0.5$ and radius up to $a=20$. 
<br>
Assuming a null pressure, the particles start to move (initially they have $velocity=0$) due to their *reciprocal gravitational potential* and then collapse into the total center of mass. 
<br>
In this program we want to build a spherical distribution of particles, so we have to generate them randomly, considering what we want to achieve. So, we need to put ourselves into a spherical coordinate frame and to think about how $r, \theta, \phi$ are distributed.
<br>
First: for *r*, the pdf distribution is 
$p(r)\propto r^2$, integrating and finding the inverse, we recover that $r\propto P^{\tfrac{1}{3}}$, where *P* is uniformly distributed.
<br>
Then for $\theta, \phi$ we have assumed to have a random uniform distribution because the object is spherically symmetric. 

```python
# generating initial points
N = int(3e2)

a = 20 #sphere radius

I = lambda x: a*x**(1/3)
P = np.random.uniform(0,1,N)

r = I(P)
theta = np.random.uniform(0, np.pi,N)
phi = np.random.uniform(0, 2*np.pi, N)

plt.hist(r, density = True, histtype = 'step', color='royalblue')
plt.ylim([0,.5]);
```

In this part we have plotted the distribution for 300 and 100000 particles to control if we are computing a spherically symmetric distribution.

```python
x = r*np.sin(theta)*np.cos(phi)
y = r*np.sin(theta)*np.sin(phi)
z = r*np.cos(theta)

N_m = int(1e5)
P_m = np.random.uniform(0,1,N_m)

r_m = I(P_m)
theta_m = np.random.uniform(0,np.pi,N_m)
phi_m = np.random.uniform(0, 2*np.pi, N_m)


x_m = r_m*np.sin(theta_m)*np.cos(phi_m)
y_m = r_m*np.sin(theta_m)*np.sin(phi_m)
z_m = r_m*np.cos(theta_m)


fig, axs = plt.subplots(1,2, figsize = (7,7), tight_layout = True, subplot_kw={'projection': '3d'})
axs[0].set(aspect='equal', title = f'{N} particles')
axs[0].scatter(x,y,z, s = .5, color='royalblue')
axs[1].set(aspect='equal', title = f'{N_m} particles')
axs[1].scatter(x_m,y_m,z_m, s =.5, color='royalblue');
```

Then, to be sure that we are doing everything correctly, we want to check if the density is constant everywhere in the sphere, so we took 3000 different radii (between 2 and 20), and we computed the density in each one.


```python
#Checking density
steps = 3000
scan = np.linspace(2, a, steps)
volume = 0
density = []

for i in scan:
    volume = (4/3) * np.pi * (i**3)
    mask = tuple([r_m <= i])
    new_r = r_m[mask]
    number = len(new_r)
    density.append(number/volume)

density = np.array(density)
counts, bins = np.histogram(density)
plt.hist(density, histtype = 'step', bins=20)
plt.xlim([2,4])
max = np.max(counts)
prob = max/steps
print('Density is compatible with a delta function for {:.2f}%.'.format(prob*100));
```

### Trying a different theta distribution

What if i consider the solid angle 

$$ \Omega = 2\sin(\theta)d\theta d\phi $$

Then the cdf will be 

$$ \text{cdf}(\theta) = -\dfrac{\cos(\theta)}{2} $$

Where i normalized the value between o and $\pi$. Then the inverse will be

$$ I(\theta) = \arccos(-2P) $$

Now we generate P with random uniform distribution

```python
# generating initial points
N = int(3e2)

a = 20 #sphere radius

I = lambda x: a*x**(1/3)
I_theta = lambda x: np.arccos(-2*x) 

P = np.random.uniform(0,1,N)
r = I(P)
theta = I_theta(P)
phi = np.random.uniform(0, 2*np.pi, N)
x = r*np.sin(theta)*np.cos(phi)
y = r*np.sin(theta)*np.sin(phi)
z = r*np.cos(theta)

N_m = int(1e5)
P_m = np.random.uniform(0,1,N_m)
r_m = I(P_m)
theta_m = I_theta(P_m)
phi_m = np.random.uniform(0, 2*np.pi, N_m)
x_m = r_m*np.sin(theta_m)*np.cos(phi_m)
y_m = r_m*np.sin(theta_m)*np.sin(phi_m)
z_m = r_m*np.cos(theta_m)


fig, axs = plt.subplots(1,2, figsize = (7,7), tight_layout = True, subplot_kw={'projection': '3d'})
axs[0].set(aspect='equal', title = f'{N} particles')
axs[0].scatter(x,y,z, s = .5, color='royalblue')
axs[1].set(aspect='equal', title = f'{N_m} particles')
axs[1].scatter(x_m,y_m,z_m, s =.5, color='royalblue');
```

# Analysis

### Free Fall and Dynamical time

We computed the free fall time for the particles that were situated at almost a radius of 20 at $t=0$, that collapsed into the center of mass. Due to the fact that the only acting force that we consider is gravity, we will have a *free fall time* of:
<br>
$$t_{ff}=\sqrt{\dfrac{3 \pi}{32G \rho}}$$

In internal units we have $G=1$ and $\rho$ is the mean over the volumes calculated previously. In cgs we find that the value amounts to $t_{ff} \approx 1.2 \text{yr}$ which can be checked with our code since the period of our Earth-Sun system used for the conversion is $T = 1 \text{yr}$ so $[t] = T/2\pi$.

The dashed vertical line represents the *theoretical free fall time*.

```python
r = np.sqrt(x**2 + y**2 + z**2)
r_i = r[:,0]                                 #i take all points at time 0
mask = (r_i >= 19.8) & (r_i <= 20.2)         #i take only the external ones at time 0


x_cm = np.sum(x[:, 0]*m)/np.sum(m)
y_cm = np.sum(y[:, 0]*m)/np.sum(m)
z_cm = np.sum(z[:, 0]*m)/np.sum(m)

r_cm = np.sqrt((x - x_cm)**2 + (y - y_cm)**2 + (z -z_cm)**2)

r_ext_cm = r_cm[mask, :]                           #i take the external ones at all times 
r_ext = r[mask, :]                           #i take the external ones at all times 

#r[0,:] is the same particle at all times 
#r[:,0] is all particles at a specific time
num_points = len(r_i[mask])

for i in range(num_points):
    plt.plot(t, r_ext[i,:],  c='royalblue', lw = .8)        #Not corrected
    plt.plot(t, r_ext_cm[i,:],  c='indianred', lw = .8)     #Corrected for CM

rho = np.sum(m)/((4/3)*np.pi*a**3)
G = 1
t_ff = np.sqrt((3*np.pi)/(32*G*rho))
print(t_ff)

plt.axvline(t_ff, color = 'black', ls = '--', lw = .8, label=f'FreeFall time = {round(t_ff, 2)}')
plt.legend()
plt.show()
```

In the end we have calculated the free fall time and the dynamical time of every particle of our distribution and confronted them with the time scale of collapse of every particle. We can see that most of the particles collapse between the free fall time and the dynamical time.
Afterwards, some of the particles run away from the others because they have a velocity that is more than the escape velocity. While the remaining form a structure that seems to be bound.

```python
for i in range(len(r)):
    plt.plot(t,r[i,:], c='royalblue', lw = .5);
rho = np.sum(m)/((4/3)*np.pi*a**3)
G = 1
t_ff = np.sqrt((3*np.pi)/(32*G*rho))
t_dyn = np.sqrt(3*np.pi/(16*G*rho))
print(f'free fall time: {round(t_ff,2)}')
print(f'dynamical time: {round(t_dyn,2)}')

plt.axvline(t_ff, color = 'black', ls = '--', lw = .8);
plt.axvline(t_dyn, color = 'black', ls = '--', lw = .8);
```
